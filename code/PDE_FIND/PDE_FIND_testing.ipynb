```python
from PDE_FIND import *
data_path = "../../datasets"
results_path = ""
import pandas as pd
import pickle
```

```python
# Simple PDE ut = 0.8ux

pred_coefficients=[]
snrs = ["No noise",20000,15000,10000,7500,5000,3000,2000,1000,800,500,300,200,100,90,80,70,60,50,40,30,20,10,8,6,5,4,3,2,1,0.9,0.8,0.75,0.7,0.6,0.5,0.4,0.3,0.25,0.2,0.1]


# Without noise: Use Finite difference method

print("="*100)
print("SNR = No noise ")
file_to_read = open(data_path+"/4_"+str(0)+".pkl", "rb")
loaded_dictionary = pickle.load(file_to_read)
u = loaded_dictionary["u"].T
x = loaded_dictionary["x"]
t = loaded_dictionary["t"]
dx = x[2]-x[1]
dt = t[2]-t[1]

Ut, R, descr = build_linear_system(u, dt, dx, D=3, P=2, time_diff = 'FD', space_diff = 'FD')
w = TrainSTRidge(R,Ut,10**-5,0.1)
print("\nPredicted PDE: ", return_pde(w, descr))
w=[np.real(el[0]) for el in w]
pred_coefficients.append([snrs[0]]+list(w))


# With noise: polynomial interpolation

for j in range(1,len(snrs)):
    print("="*100)
    print("SNR = ",snrs[j])

    # Loading u,x,t
    file_to_read = open(data_path+"/4_"+str(j)+".pkl", "rb")
    loaded_dictionary = pickle.load(file_to_read)
    u = loaded_dictionary["u"].T
    x = loaded_dictionary["x"]
    t = loaded_dictionary["t"]
    dx = x[2]-x[1]
    dt = t[2]-t[1]
    
    Ut, R, descr = build_linear_system(u, dt, dx, D=3, P=2, time_diff = 'poly',
                                            deg_x = 4, deg_t = 4, 
                                            width_x = 10, width_t = 10)
    w = TrainSTRidge(R,Ut,10**-5,1)
    print("\nPredicted PDE: ", return_pde(w, descr))
    w=[np.real(el[0]) for el in w]
    pred_coefficients.append([snrs[j]]+list(w))
    
df_4 = pd.DataFrame(pred_coefficients, columns = ["SNR"]+descr )
df_4 = df_4[['SNR', '', 'u_{x}', 'u_{xx}', 'u_{xxx}', 'u',
       'u^2', 'uu_{x}', 'u^2u_{x}', 'uu_{xx}', 'u^2u_{xx}', 'uu_{xxx}',
       'u^2u_{xxx}']]
df_4.to_csv(results_path+"/simplepde_results.csv")
```
