```python
from GKFL0opt import *
data_path = "../../datasets"
results_path = "../KNNL0"
import pandas as pd
```

```python
# Simple PDE ut = 0.8ux

goodsigmas_mp={}
bestsigmas_mp={}


actual_eqn=[0]*12
actual_eqn[1]=-0.8
    
pred_coefficients=[]


snrs = ["No noise",20000,15000,10000,7500,5000,3000,2000,1000,800,500,300,200,100,90,80,70,60,50,40,30,20,10,8,6,5,4,3,2,1,0.9,0.8,0.75,0.7,0.6,0.5,0.4,0.3,0.25,0.2,0.1]

for j in range(len(snrs)):
    print("="*100)
    print("SNR = ",snrs[j])

    # Loading u,x,t
    path = data_path+"/4_"+str(j)+".pkl"
    
    # Finding best sigma
    goodsigmas,bestsigma = gkfl0opt(path=path,actual=actual_eqn,correct_id='010000000000',sigmastart=0,sigmaend=15,nsigma=31)
    print("\nGoodsigmas: ",goodsigmas)
    print("Bestsigma: ",bestsigma)
    goodsigmas_mp[snrs[j]]=goodsigmas
    bestsigmas_mp[snrs[j]]=bestsigma
    
    # Using best sigma
    u,x,t = load_data(path)
    sols,descr = discover_pde(u,x,t,sigma=bestsigma)
    print("\nPDEs Predicted by GKFL0opt: ")
    for i in range(len(sols)):
        print(return_pde(sols[i]["B"], descr))
    
    # Desired PDE
    w= sols[0]["B"]
    print("\nDesired PDE: ", return_pde(w, descr))
    pred_coefficients.append([snrs[j]]+list(w))
    
df_4 = pd.DataFrame(pred_coefficients, columns = ["SNR"]+descr )
df_4.to_csv(results_path+"/simplepde_results.csv")
print(goodsigmas_mp)
print(bestsigmas_mp)
```
