```python
from GKFL0 import *
data_path = "../../datasets"
results_path = "../GKFL0"
import pandas as pd
```

```python
# Simple PDE ut = 0.8ux

pred_coefficients=[]

snrs = ["No noise",20000,15000,10000,7500,5000,3000,2000,1000,800,500,300,200,100,90,80,70,60,50,40,30,20,10,8,6,5,4,3,2,1,0.9,0.8,0.75,0.7,0.6,0.5,0.4,0.3,0.25,0.2,0.1]

for j in range(len(snrs)):
    print("="*100)
    print("SNR = ",snrs[j])

    # Loading u,x,t
    path = data_path+"/4_"+str(j)+".pkl"
    sols,descr = discover_pde(path,sigma=6)
    print("\nPDEs Predicted by GKFL0: ")
    for i in range(len(sols)):
        print(return_pde(sols[i]["B"], descr))
    
    # Desired PDE
    w= sols[0]["B"]
    print("\nDesired PDE: ", return_pde(w, descr))
    pred_coefficients.append([snrs[j]]+list(w))
    
df_4 = pd.DataFrame(pred_coefficients, columns = ["SNR"]+descr )
df_4.to_csv(results_path+"/simplepde_results.csv")
```

```python
def plot(u,x,t,title="u vs x and t",size_of_plot=(8, 5),font_size=10):
    """ 
    Plots u vs x and t
    """
    X, T = np.meshgrid(x, t)
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    surf = ax.plot_surface(X, T, u, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm,linewidth=0, antialiased=False)
    plt.title(title)
    plt.xlabel('x', fontsize = font_size)
    plt.ylabel('t', fontsize = font_size)
    ax.set_zlim(-0.4, 1)
    plt.show()
```
